```python
section_names = np.unique(np.array(data['section_name']))
print section_names
```

```python
def create_subset(section_names, num):
    subset = pd.DataFrame()
    for name in section_names:
        new_rows = data[data['section_name']==name] 
        if new_rows.shape[0] > num:
            new_rows = new_rows.iloc[:num,] #gives 5 instances vs loc which is specific to loc
        subset = pd.concat([subset, new_rows])
    return subset

subset = create_subset(section_names, 5)
print subset.shape
# grab section
# grab 5 instances of each section
```

```python
def create_bag(df, vectorizer): 
    features = df['content']
    bag_words = vectorizer.fit_transform(features)
    return bag_words
```
